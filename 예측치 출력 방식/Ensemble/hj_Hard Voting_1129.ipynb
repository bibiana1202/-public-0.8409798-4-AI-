```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score
import os 
```

```python
def mode(list):
    count = 0
    mode = 0;
    for x in list:
      if list.count(x) > count:
          count = list.count(x)
          mode = x

    return mode
```

```python
import os
xlsx_list = [file for file in os.listdir('/content') if file.endswith('csv')]
xlsx_list
```

```python
len(xlsx_list)
```

```python
y_predict = []
for csv_file in xlsx_list:
  y_predict.append(pd.read_csv(f'/content/{csv_file}').N_category.values)
```

```python
prediction = []

for i in range(250):
  predict = []
  for j in y_predict:
    if j[i] > 0.4:
      predict.append(1)
    else:
      predict.append(0)
  prediction.append(mode(predict))
```

```python
submit = pd.read_csv('/content/sample_submission.csv')
```

```python
submit['N_category'] = prediction
submit.to_csv('./hard_voting_submit.csv', index=False)
```
